# Populate frame split csv 

```python
import csv
import os
import pdb

csv_path = "/n/holylabs/gershman_lab/Users/zkelso/video_splits.csv"          # path to your CSV
folder_path = "/n/netscratch/gershman_lab/Lab/zkelso/GIFs/"    # folder with GIF (or other) files

SUFFIX = "_fullvideo_points_GIF.gif"

# 1. Read existing CSV names
existing_rows = []
existing_names = set()

if os.path.exists(csv_path):
    with open(csv_path, newline="") as f:
        reader = csv.reader(f)
        for row in reader:
            if not row:
                continue
            name = row[0].strip()
            existing_rows.append(row)
            existing_names.add(name)

# 2. Scan folder and prepare new rows
new_rows = []

for fname in os.listdir(folder_path):
    # ignore non-matching names if needed
    if not fname.endswith(SUFFIX):
        continue

    # strip suffix to get the CSV-style name
    base_name = fname[: -len(SUFFIX)]

    if base_name not in existing_names:
        new_rows.append([base_name, "[]"])
        existing_names.add(base_name)

pdb.set_trace()
# 3. Rewrite CSV with old + new rows
if new_rows:
    with open(csv_path, "w", newline="") as f:
        pdb.set_trace()
        writer = csv.writer(f)
        # write header if you have one:
        # writer.writerow(["name", "frame_splits"])

        for row in existing_rows:
            writer.writerow(row)
        for row in new_rows:
            writer.writerow(row)
```
